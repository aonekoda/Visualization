# 범주형 데이터의 산점도, Categorical scatterplots
범주별로 산점도를 표현할 때 사용된다.

* stripplot() (with kind="strip"; the default) - 한 쪽 변수가 범주형인 산점도를 그린다.
* swarmplot() (with kind="swarm") - stripplot()과 유사하지만 점이 겹치지 않도록 조정된다.(범주형 축을 따라서).

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
tips = pd.read_csv('./data/tips.csv')
tips.info()
````

````python
titanic = pd.read_csv('./data/titanic.csv')
titanic.info()
````

## Stripplot

범주별 데이터 포인트들을 일일이 점 형태로 찍어서 표시한다.

````python
sns.stripplot(x='Pclass', y='Age', data=titanic)
````

````python
sns.stripplot(x='Pclass', y='Age', hue='Sex', alpha=0.5, data=titanic)
````

````python
ax = sns.stripplot(x="day", y="total_bill", data=tips)
````

````python
# jitter 분량을 조정한다.
ax = sns.stripplot(x="day", y="total_bill", data=tips, jitter=0.05)
````

````python
ax = sns.stripplot(x="total_bill", y="day", hue='sex', data=tips)
````

## Swarmplot
데이터의 분산까지 고려하여, 데이터 포인트가 서로 중복되지 않도록 그린다. 

````python
plt.figure(figsize=(10,5))
sns.swarmplot(x='Pclass', y='Age', data=titanic)
````

````python
ax = sns.swarmplot(x="day", y="total_bill", data=tips)
````

````python
ax = sns.swarmplot(x="day", y="total_bill", hue="sex", data=tips)
````

## stipplot vs swarmplot

* stripplot : 데이터 포인트가 중복되어 범주별 분포를 그린다.
* swarmplot : 위 그래프는 데이터의 분산까지 고려하여, 데이터 포인트가 서로 중복되지 않도록 그린다. 즉, 데이터가 퍼져 있는 정도를 입체적으로 볼 수 있다.

````python
fig, axes = plt.subplots(1,2, figsize=(20,5))

ax1 = sns.stripplot(x='Pclass', y='Age', hue='Sex', data=titanic, ax=axes[0])
ax1.set_title('Stripplot')

ax2 = sns.swarmplot(x='Pclass', y='Age', hue='Sex', data=titanic, ax=axes[1])
ax2.set_title('Swarmplot')
````

### sina plot
box plot이나 violin plot 에 swarmplot을 겹쳐서 그리면 데이터의 분포를 더 자세히 표현할 수 있다.


````python
plt.figure(figsize=(10,5))
sns.boxplot(x='Pclass', y='Age', data=titanic)
sns.swarmplot(x='Pclass', y='Age', data=titanic, color=".5")
````
